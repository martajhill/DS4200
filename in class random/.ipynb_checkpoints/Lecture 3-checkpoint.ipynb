### Describing a data set

In a data table (also called ``data matrix"), like the one below. There are two dimensions in the data: 

- Observations: each individual subject, case, or sample point (rows, represented with $n$)
- Variables/Feature: characteristics that are recorded for each subject in the sample (columns, represented with $p$)


```python
import pandas as pd

df = pd.read_csv("ICUAdmissions.csv")
print(df.head())
```

### Describing variables

There are at least two ways to describe variables:
		
By value type:
		
- Numerical variables: take on numerical values with meaning Mathematical operations like addition, etc. must make sense
- Categorical variables: take on names, categories, or labels Categories are sometimes represented by numbers...

By function:
		
- Response variable/Target: defined by the particular research question a study seeks to address, and measures the outcome of interest in the study
- Explanatory variable/Feature: a secondary variable(s) that explains changes in the response variable

### Basic data visualization choices

Here is a general guidline for the type of plots: 

- One categorical: bar plot
- One continous: histogram/density plot/boxplot
- Two continous: scatter plot
- One categorical and one continous: side-by-side boxplot
- Two categorical: side-by-side bar plot 

#### One categorical: bar plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x = "Status", data = df)
plt.xlabel("Status")
plt.ylabel("Count")
plt.title("Count of status in ICU admission")


status_counts = df["Status"].value_counts()
for i, value in enumerate(status_counts.values):
    plt.text(i, value+1, f'{value}', ha = 'center')
```

```python
status_counts = df["Status"].value_counts(normalize = True)*100

sns.barplot(x = status_counts.index, y = status_counts.values)
plt.xlabel("Status")
plt.ylabel("Percentage")
plt.title("Status percentage in ICU admission")

for i, value in enumerate(status_counts.values):
    plt.text(i, value+1, f'{value:.2f}%', ha ='center')
```

##### What to write about a bar plot
- What is this figure about?
- Percetage/count for each group
- whether there is a clear difference between groups
This figure shows the number of people who survive an ICU admission. Status '0' means the patient survived and status '1' means the patient died. Among the 200 patients in the data set, 40 patients (20%) died and the remaining 160 (80%) survived. It seems there is a difference between the two groups. 

#### One continous: histogram/density plot/boxplot

```python
sns.histplot(df["HeartRate"],kde = True, bins = 20, kde_kws = {'bw_method': 0.6})
plt.xlabel('Heart Rate')
plt.ylabel('Frequency and Density')
plt.title('Heart rate density in ICU admissions')
```

```python
sns.kdeplot(df["HeartRate"], fill = True, bw_adjust=0.6)
plt.xlabel('Heart Rate')
plt.ylabel('Frequency and Density')
plt.title('Heart rate density in ICU admissions')
```

```python
sns.boxplot(df["HeartRate"])
plt.xlabel('Heart Rate')
plt.ylabel('Frequency and Density')
plt.title('Heart rate density in ICU admissions')
```

##### Difference between histogram/density plot/ boxplot

- Histogram and density plot emphasize distribution of data
- Box plot emphasizes statistics of the data (median, IQR)
- Box plot does not catch distirbution information (like bi-mode)
- For histogram you need to choose the bin number and density plot you need to choose the bandwidth
- ^ both dramatically affect the plot

##### Difference between histogram and bar plot

- Histogram is for numerical data and bar plot is for categorical data
- Histogram for the most case is vertical but bar plot can be horizontal
- Bars can switch order in bar plot, can't in histogram
-


##### What to write about a histogram/density plot

- shape (skewness and mode)
- majorityof data (Typical value: mean, pm sd)
- Potential outlier

example: In the figure, the heart rate has a uni-model distributionw hich is slightly skewed to the right (positive skew). The typical value is within the range $98.9 \pm 26.8 = [72.1, 125.7]$. there might exist an outlier with large value.

##### What to write about a box plot

- still want to comment on the skew but not the shape
- Majority of the data (IQR)
- potential outliers

example: In the figure, the heart rate has a distribution slightly skewed to the right. The IQR range is from 80 to 118.25. There may exist a large outlier.

#### Two continous: scatter plot

```python
plt.scatter(df["Age"], df["HeartRate"], alpha = 0.7)
plt.xlabel("Age")
plt.ylabel("Heart Rate")
plt.title("Scatter plot of age and heart rate in ICU admission data")
```

##### What to write about a scatter plot

- trend (linear or nonlinear)
- are there any outliers?
- Are there any clusters?
Example: In the figure there is no clear trend in the relationship between age and heart rate. There are no outlying data points, all points are near by. No patter of clusters emerges. 


#### One categorical and one continous: side-by-side boxplot

Other choices include side-by-side histogram and side-by-side density plot

```python
sns.boxplot(x = "Status", y = 'HeartRate', data = df)
plt.xlabel("Status")
plt.ylabel("Heart Rate")
plt.title("Heart rate by status in ICU admissions")
```

##### What to write about a side-by-side boxplot or density plot

- highlight comparison
- for boxplot: declare whether the boxes overlap (vertically, not literally)

Example: in the figure, the boxplots overlap, which means there is no practical differences in heart rate of the group that survived and the group that died. 


- statistical difference: usually indicating that a statistics test was run to show this. (T test with small P value)
- practical difference: something you can see in data or data visualization

#### Two categorical: side-by-side bar plot

```python
sns.countplot(x = "Status", hue = "CPR", data = df)

plt.xlabel("Status")
plt.ylabel("Count")
plt.title("Side-by-side bar plot for CPR and status")
```

##### What to write about a side-by-side bar plot

- highlight comparison

Ex: in the figure, the distributino of CPR is different for the survival group and the group that did not survive. In the survival group, there is a much larger count of patients who do not use CPR, compared to the group that did not survive. This indicated there may exist a relationship between whether CPR is used and the survival chances.
